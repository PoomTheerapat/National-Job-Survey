```python
before = 0
after = 0

for i in range(0, 10):
    print(f'--- {i} ปี ---')
    
    # Skip 6 years
    if i == 6:
        continue
    
    condition = (
        (massage["อาชีพในปัจจุบัน ท่านทำมานานเท่าใด"] == i) |
        (massage["อาชีพในปัจจุบัน ท่านทำมานานเท่าใด"] == f'{i}ปี') |
        (massage["อาชีพในปัจจุบัน ท่านทำมานานเท่าใด"] == f'{i} ปี') |
        (massage["อาชีพในปัจจุบัน ท่านทำมานานเท่าใด"] == f'{i}ปี6เดือน') |
        (massage["อาชีพในปัจจุบัน ท่านทำมานานเท่าใด"] == f'{i}ป')
    )
    
    # Count the occurrences based on the condition
    count = massage.loc[condition, 'อาชีพในปัจจุบัน ท่านทำมานานเท่าใด'].value_counts().sum()
    before += count
    
    print(massage.loc[condition, 'อาชีพในปัจจุบัน ท่านทำมานานเท่าใด'].value_counts())
    print(count)
    
    before = before + massage[condition][['อาชีพในปัจจุบัน ท่านทำมานานเท่าใด']].value_counts().sum()
    
    # Update the DataFrame based on the condition
    massage.loc[condition, 'อาชีพในปัจจุบัน ท่านทำมานานเท่าใด'] = i
    
    after = after + massage[condition][['อาชีพในปัจจุบัน ท่านทำมานานเท่าใด']].value_counts().sum()

print()
print('SUM Before: ', before)
print('SUM After: ', after)
```
