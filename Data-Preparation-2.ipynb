### Change Columns Type

```python
data['เพศ'] = data['เพศ'].astype(str)
data['จำนวนคนที่ร่วมใช้จ่ายจากรายได้ของท่าน'] = data['จำนวนคนที่ร่วมใช้จ่ายจากรายได้ของท่าน'].astype(int)
data['รายได้รวมเฉลี่ยของท่านในปัจจุบัน'] = data['รายได้รวมเฉลี่ยของท่านในปัจจุบัน'].astype(int)
data['รายได้ที่ได้รับจากการทำงาน'] = data['รายได้ที่ได้รับจากการทำงาน'].astype(int, errors='ignore')
```

```python
data['ไม่ได้รับค่าตอบแทนตามที่ตกลงไว้'] = data['ไม่ได้รับค่าตอบแทนตามที่ตกลงไว้'].astype(str)
data['ได้รับค่าตอบแทนล่าช้า'] = data['ได้รับค่าตอบแทนล่าช้า'].astype(str) 
data['ชั่วโมงทำงานมากเกินไป'] = data['ชั่วโมงทำงานมากเกินไป'].astype(str)
data['ไม่สามารถหยุดงานได้'] = data['ไม่สามารถหยุดงานได้'].astype(str)
data['ไม่มีสวัสดิการ'] = data['ไม่มีสวัสดิการ'].astype(str)
```

## Dummy Variable

```python
data.loc[data['ไม่ได้รับค่าตอบแทนตามที่ตกลงไว้'] == '0', 'ไม่ได้รับค่าตอบแทนตามที่ตกลงไว้'] = 'ไม่ใช่'
data.loc[data['ไม่ได้รับค่าตอบแทนตามที่ตกลงไว้'] == '1', 'ไม่ได้รับค่าตอบแทนตามที่ตกลงไว้'] = 'ใช่'
data.loc[data['ได้รับค่าตอบแทนล่าช้า'] == '0', 'ได้รับค่าตอบแทนล่าช้า'] = 'ไม่ใช่'
data.loc[data['ได้รับค่าตอบแทนล่าช้า'] == '1', 'ได้รับค่าตอบแทนล่าช้า'] = 'ใช่'
data.loc[data['ชั่วโมงทำงานมากเกินไป'] == '0', 'ชั่วโมงทำงานมากเกินไป'] = 'ไม่ใช่'
data.loc[data['ชั่วโมงทำงานมากเกินไป'] == '1', 'ชั่วโมงทำงานมากเกินไป'] = 'ใช่'
data.loc[data['ไม่สามารถหยุดงานได้'] == '0', 'ไม่สามารถหยุดงานได้'] = 'ไม่ใช่'
data.loc[data['ไม่สามารถหยุดงานได้'] == '1', 'ไม่สามารถหยุดงานได้'] = 'ใช่'
data.loc[data['ไม่มีสวัสดิการ'] == '0', 'ไม่มีสวัสดิการ'] = 'ไม่ใช่'
data.loc[data['ไม่มีสวัสดิการ'] == '1', 'ไม่มีสวัสดิการ'] = 'ใช่'
```

```python
# Gender
data.loc[data['เพศ'] == '1', 'เพศ'] = 'ชาย'
data.loc[data['เพศ'] == '2', 'เพศ'] = 'หญิง'
```
