# Provinces Surveyed

```python
# Assuming massage is your DataFrame
top_3_provinces = data['จังหวัดที่สำรวจ'].value_counts().nlargest(10).index

plt.figure(figsize=(10, 5))

# Filter the data for the top 3 provinces
filtered_data = data[data['จังหวัดที่สำรวจ'].isin(top_3_provinces)]

plt.hist(filtered_data['จังหวัดที่สำรวจ'], bins=10, edgecolor='black', color='blue')  # You can adjust the number of bins as needed
plt.xlabel('Provinces')

counts, bins, _ = plt.hist(filtered_data['จังหวัดที่สำรวจ'], bins=10, edgecolor='black')  

# Calculate midpoints of each bin
midpoints = (bins[:-1] + bins[1:]) / 2
plt.xticks(midpoints)  # Set x-ticks at midpoints
plt.ylabel('Frequency')
plt.title('Top 10 Provinces Surveyed The Most Massage Job')

# Add text annotations on top of each bar
for count, x in zip(counts, midpoints):
    plt.text(x, count, str(int(count)), ha='center', va='bottom')

plt.show()
```

```python
# Assuming 'data' is your DataFrame
for i in data['จังหวัดที่สำรวจ'].unique():
    mean_income = data[data['จังหวัดที่สำรวจ'] == i]['รายได้รวมเฉลี่ยของท่านในปัจจุบัน'].mean()
    print(f'จังหวัดที่สำรวจ: {i}')
    print(f'Mean Income: {mean_income}')
    print('-----------')

# Creating the graph
data_grouped = data.groupby('จังหวัดที่สำรวจ')['รายได้รวมเฉลี่ยของท่านในปัจจุบัน'].mean()
data_grouped.plot(kind='bar')

# Annotating each bar with its value
for index, value in enumerate(data_grouped):
    plt.text(index, value, str(round(value, 2)), ha='center', va='bottom')

plt.xlabel('จังหวัดที่สำรวจ')
plt.ylabel('รายได้เฉลี่ย')
plt.title('รายได้เฉลี่ยแยกตามจังหวัดที่สำรวจ')
plt.show()
```

# ไม่มีสวัสดิการ

```python
data['ไม่มีสวัสดิการ'].value_counts()
```

```python
for i in data['ไม่มีสวัสดิการ'].unique():
        
        print(f'ประเภท: {i}')
        print(data[data['ไม่มีสวัสดิการ'] == i]['รายได้รวมเฉลี่ยของท่านในปัจจุบัน'].describe())
        print('-----------', end='\n\n')
```

```python
# Assuming 'data' is your DataFrame
for i in data['ไม่มีสวัสดิการ'].unique():
    mean_income = data[data['ไม่มีสวัสดิการ'] == i]['รายได้รวมเฉลี่ยของท่านในปัจจุบัน'].mean()
    print(f'ไม่มีสวัสดิการ: {i}')
    print(f'Mean Income: {mean_income}')
    print('-----------')

# Creating the graph
data_grouped = data.groupby('ไม่มีสวัสดิการ')['รายได้รวมเฉลี่ยของท่านในปัจจุบัน'].mean()
data_grouped.plot(kind='bar')

# Annotating each bar with its value
for index, value in enumerate(data_grouped):
    plt.text(index, value, str(round(value, 2)), ha='center', va='bottom')

plt.xlabel('ไม่มีสวัสดิการ')
plt.ylabel('รายได้เฉลี่ย')
plt.title('ไม่มีสวัสดิการและรายได้เฉลี่ย')
plt.show()
```

# ท่านอยากทำงานในระบบหรือองค์กรณ์ที่มีระบบสวัสดิการหรือไม่

```python
data['ท่านอยากทำงานในระบบหรือองค์กรณ์ที่มีระบบสวัสดิการหรือไม่'].value_counts()
```

```python
# Assuming 'data' is your DataFrame
for i in data['ท่านอยากทำงานในระบบหรือองค์กรณ์ที่มีระบบสวัสดิการหรือไม่'].unique():
    mean_income = data[data['ท่านอยากทำงานในระบบหรือองค์กรณ์ที่มีระบบสวัสดิการหรือไม่'] == i]['รายได้รวมเฉลี่ยของท่านในปัจจุบัน'].mean()
    print(f'ท่านอยากทำงานในระบบหรือองค์กรณ์ที่มีระบบสวัสดิการหรือไม่: {i}')
    print(f'Mean Income: {mean_income}')
    print('-----------')

# Creating the graph
data_grouped = data.groupby('ท่านอยากทำงานในระบบหรือองค์กรณ์ที่มีระบบสวัสดิการหรือไม่')['รายได้รวมเฉลี่ยของท่านในปัจจุบัน'].mean()
data_grouped.plot(kind='bar')

# Annotating each bar with its value
for index, value in enumerate(data_grouped):
    plt.text(index, value, str(round(value, 2)), ha='center', va='bottom')

plt.xlabel('ท่านอยากทำงานในระบบหรือองค์กรณ์ที่มีระบบสวัสดิการหรือไม่')
plt.ylabel('รายได้เฉลี่ย')
plt.title('ท่านอยากทำงานในระบบหรือองค์กรณ์ที่มีระบบสวัสดิการหรือไม่และรายได้เฉลี่ย')
plt.show()
```
