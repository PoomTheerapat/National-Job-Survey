# Provinces Surveyed

```python
# Assuming massage is your DataFrame
top_3_provinces = data['จังหวัดที่สำรวจ'].value_counts().nlargest(10).index

plt.figure(figsize=(10, 5))

# Filter the data for the top 3 provinces
filtered_data = data[data['จังหวัดที่สำรวจ'].isin(top_3_provinces)]

plt.hist(filtered_data['จังหวัดที่สำรวจ'], bins=10, edgecolor='black')  # You can adjust the number of bins as needed
plt.xlabel('Provinces')

counts, bins, _ = plt.hist(filtered_data['จังหวัดที่สำรวจ'], bins=10, edgecolor='black')  

# Calculate midpoints of each bin
midpoints = (bins[:-1] + bins[1:]) / 2
plt.xticks(midpoints)  # Set x-ticks at midpoints
plt.ylabel('Frequency')
plt.title('Top 10 Provinces Surveyed The Most Message Job')

# Add text annotations on top of each bar
for count, x in zip(counts, midpoints):
    plt.text(x, count, str(int(count)), ha='center', va='bottom')

plt.show()
```

# เพศ

```python
data['เพศ'].value_counts()
```

```python
plt.figure(figsize=(6, 5))  # Set the figure size (width, height)
hist_values, bins, _ = plt.hist(data['เพศ'], bins=2, edgecolor='black', align='mid')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Gender Distribution')

# Manually set x-ticks to the midpoints of the bins
plt.xticks([(bins[i] + bins[i+1]) / 2 for i in range(len(bins)-1)], ['หญิง', 'ชาย'])

# Add numbers on top of each bar
for i, count in enumerate(hist_values):
    plt.text((bins[i] + bins[i+1]) / 2, count, str(int(count)),
             ha='center', va='bottom')

plt.show()
```

# อายุ

```python
data['อายุ'].describe()
```

```python
hist_values, bins, _ = plt.hist(data['อายุ'], bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

# Add numbers in the center of each bar
for i, value in enumerate(hist_values):
    plt.text(bins[i] + (bins[i+1] - bins[i]) / 2, value, str(int(value)),
             ha='center', va='bottom')

plt.show()
```

```python
# Extract unique values and their counts
unique_values, counts = data['อายุ'].value_counts().sort_index().index, data['อายุ'].value_counts().sort_index().values

# Plotting the histogram with discrete values
plt.bar(unique_values, counts, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

# Add numbers on top of each bar
for i, count in enumerate(counts):
    plt.text(unique_values[i], count, str(int(count)),
             ha='center', va='bottom')

plt.show()
```
