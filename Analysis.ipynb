### จังหวัดที่สำรวจ

```python
massage['จังหวัดที่สำรวจ'].value_counts(sort = True)
```

```python
# Assuming massage is your DataFrame
top_3_provinces = massage['จังหวัดที่สำรวจ'].value_counts().nlargest(10).index

plt.figure(figsize=(10, 5))

# Filter the data for the top 3 provinces
filtered_data = massage[massage['จังหวัดที่สำรวจ'].isin(top_3_provinces)]

plt.hist(filtered_data['จังหวัดที่สำรวจ'], bins=10, edgecolor='black')  # You can adjust the number of bins as needed
plt.xlabel('Provinces')

counts, bins, _ = plt.hist(filtered_data['จังหวัดที่สำรวจ'], bins=10, edgecolor='black')  

# Calculate midpoints of each bin
midpoints = (bins[:-1] + bins[1:]) / 2
plt.xticks(midpoints)  # Set x-ticks at midpoints
plt.ylabel('Frequency')
plt.title('Top 10 Provinces Surveyed')

# Add text annotations on top of each bar
for count, x in zip(counts, midpoints):
    plt.text(x, count, str(int(count)), ha='center', va='bottom')

plt.show()
```

#### ส่วนใหญ่ที่สำรวจเจอเกี่ยวกับการนวดเจอที่ เชียงใหม่มากที่สุด จำนวน 21 คน
