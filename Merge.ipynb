# Explore NaN Values

```python
massage.isna().sum()
```

```python
for column in massage.columns:
    nan_count = massage[column].isna().sum()
    if nan_count > 0:        
        print(massage[column].value_counts(),end='\n\n')
        print('Sum Record: ', massage[column].value_counts().sum())
        print('Sum NaN: ', massage[column].isna().sum())
        print('Column Type: ',massage[column].dtype, end='\n\n')
        print('----------------------------')
```

```python
for column in massage.columns:      
        print(massage[column].value_counts(),end='\n\n')
        print('----------------------------')
```
